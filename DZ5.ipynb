```python
import json
dict_purchase = {}
with open('purchase_log.txt', encoding = 'utf8') as f:
    next(f)
    for line in f:
        line = line.strip()       
        dict_ = json.loads(line)
        dict_purchase.setdefault(dict_['user_id'],dict_['category'])
```

```python
import csv
f = open('visit_log.csv', 'r') # as f:
next(f)
with open('new_visit_log.csv','w', newline='', encoding = 'utf8') as out_f:
    columns = ['user_id', 'source', 'category']
    writer = csv.DictWriter(out_f, fieldnames = columns)
    writer.writeheader()
    for line in  f:
        line = line.strip().split(',')
        for k in line:
            for d in dict_purchase.items():
                if k == d[0] and line[1] != 'None':
                    new_line = {'user_id': k, 'source': line[1], 'category': dict_purchase[k]}
                    writer.writerow(new_line)
        new_line = {'user_id': line[0], 'source': line[1], 'category': ''}
        writer.writerow(new_line)
f.close()
```
